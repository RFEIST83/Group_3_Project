```python
driver_winners = pd.merge(results, drivers[['driverId', 'forename', 'surname', 'driverRef']], on='driverId')

driver_winners = pd.merge(driver_winners, constructors[['name', 'constructorId']], on='constructorId')

driver_winners['position'] = pd.to_numeric(driver_winners['position'], errors='coerce')

driver_winners['driverName'] = driver_winners['forename'] + ' ' + driver_winners['surname']

driver_winners = driver_winners.groupby(['name', 'driverRef', 'driverName'], as_index=False).agg(Wins=('position', lambda x: (x == 1).sum()))

driver_winners = driver_winners.sort_values(by='Wins', ascending=False)

print(driver_winners)
```

```python
top_5_driverNames = driver_winners.groupby('driverName')['Wins'].sum().nlargest(5).index

top_5_data = driver_winners[driver_winners['driverName'].isin(top_5_driverNames)]

for driverName in top_5_driverNames:
    driver_data = top_5_data[top_5_data['driverName'] == driverName]
    
    explode_index = driver_data['Wins'].idxmax()
    
    explode = [0.1 if i == explode_index - driver_data.index[0] else 0 for i in range(len(driver_data))]
    
    plt.figure(figsize=(8, 8))
    plt.pie(driver_data['Wins'], labels=driver_data['name'], autopct='%1.1f%%', startangle=140, explode=explode)
    plt.title(f'Winning Percentage for: {driverName}')
    plt.show()
```

```python
driver_winners = pd.merge(results, drivers[['driverId', 'forename', 'surname', 'driverRef']], on='driverId')

driver_winners = pd.merge(driver_winners, constructors[['name', 'constructorId']], on='constructorId')

driver_winners['position'] = pd.to_numeric(driver_winners['position'], errors='coerce')

driver_winners['driverName'] = driver_winners['forename'] + ' ' + driver_winners['surname']

driver_winners = driver_winners.groupby(['name', 'driverRef', 'driverName'], as_index=False).agg(
    Wins=('position', lambda x: (x == 1).sum()),
    Losses=('position', lambda x: (x != 1).sum())
)

driver_winners = driver_winners.sort_values(by='Wins', ascending=False)

print(driver_winners)
```

```python
top_5_driverNames = driver_winners.groupby('driverName')['Wins'].sum().nlargest(5).index

top_5_data = driver_winners[driver_winners['driverName'].isin(top_5_driverNames)]

def generate_pie_chart(data, driverName):
    
    plt.figure(figsize=(8, 8))
    plt.pie(
        [data['Wins'].values[0], data['Losses'].values[0]],
        labels=['Wins', 'Losses'],
        autopct='%1.1f%%',
        startangle=140,
        colors = ['lightblue', 'lightcoral'],
        explode=[0.1, 0]
    )
    plt.title(f'% of Total Wins and Losses for: {driverName}')
    plt.show()

for driverName in top_5_driverNames:
    driver_data = top_5_data[top_5_data['driverName'] == driverName]
    generate_pie_chart(driver_data, driverName)
```
